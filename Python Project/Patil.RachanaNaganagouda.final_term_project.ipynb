### The CNN Moneyâ€™s Market Movers website (https://money.cnn.com/data/hotstocks/ ) tracks the most active stocks on a real time basis. Specifically, the most active, the top gainers and top losers are listed at any instance in time. You will first write Python scripts that collect the list of most actives, gainers and losers from the above website. Next, your programs should take the ticker symbols and names of these companies (and categories) and build a csv file (called stocks.csv) with data about each stock from the website: https://finance.yahoo.com/quote/AMD?p=AMD&.tsrc=fin-srch-v1 which gives the quote for ticker symbol AMD as an example. The data to be collected from the Yahoo Finance site should include:
### OPEN price
### PREV CLOSE price
### VOLUME
### MARKET CAP
### Your code should also list the names of the companies in the order and categories listed in the website: https://money.cnn.com/data/hotstocks/ and ask the user to choose a company to get the data on. Once the user chooses the company of interest, your program should display its corresponding data (Open, Prev Close, Volume and Market Cap).

### The folllowing code works as:
### The modules requests, bs4 and csv are imported. With the help of requests, the website mentioned below is accessed. Then the content of website is parsed with help of BeautifulSoup. The table is scraped with the help of required class name. The lists such as tickersymbols,fullnames and maintypes are initialised respectively. Any empty dictionary is initialised. A variable called counter is initialised to zero. With the help of csv writer the stocks .csv file is created and opened in write mode. The csv_writer is  a pointer which will help contents to be written in the file. The table variable consists of a list which contains contents of all three tables. The contents are accessed through 'for' loop one by one. Inside table, all rows are found out in the list. Inside each row, first column is scraped out. The abbreviated company names are appended to tickersymbols list whereas full names are appended to fullnames list. Then each abbreviated company name is given to the second website in the form of link and is navigated to that company page and scraped out required values as shown below. The values written to the csv file are the type of the stock, the abbreviated name, full name, previous value, open, volume and and market cap. all the values are wriiten in the csv file as a row . The counter is incremented to pint to the different stock type and same process goes on. The file is then closed after writing. Then the file is opened  in read mode to display the comapny names for the user to check through it. The contents are displayed accordingly. The user is prompted to enter the company name as per the list. If the comapny is mentioned in the list, The details are displayed else it prints company not found. 

```python
import requests                                    # import the required modules like requests, bs4 and csv
from bs4 import BeautifulSoup
import csv

response=requests.get('https://money.cnn.com/data/hotstocks/')      # get the website content with the help of requests module
soup=BeautifulSoup(response.text,'html.parser')                     # output is parsed to get required readable content 
table=soup.find_all(class_="wsod_dataTable wsod_dataTableBigAlt")   #All table content is scraped through required class.


tickersymbols=[]
fullnames=[]                                  # required lists are initialised
types=dict()
maintypes=["Most Actives", "Gainers","Losers"]  
count=0                                             # count variable initialised to 0
           
with open('stocks.csv','w',newline='') as csv_file:      # open the stock.csv file in write mode
    
    csv_writer=csv.writer(csv_file,delimiter=",")       # initialise the pointer
    for tab in table:                                   # iterate through the table list 
        rows=tab.find_all('tr')                         # scrape all the row
        for row in rows:
            data=row.find_all('td')[0:1]                # scrape the first column of the row
            for d in data:
                t=d.find("a",attrs={'class':"wsod_symbol"}).get_text()  # scarpe the tickersymbols of comapnies
                #t=d.find(class_="wsod_symbol").get_text()
                tickersymbols.append(t)                                # append it to the tickersymbol list
                w=d.find_all('span')[0]                             # srape full name of the list
                fn=w.get_text()
                fullnames.append(w.get_text())                      # append it to the fullname list
                types[t]=maintypes[count]                          # map the tickersymbol to its typw in the dictionary
                
                tickerurl='https://finance.yahoo.com/quote/{}?p={}&.tsrc=fin-srch-v1'.format(t,t)
                response=requests.get(tickerurl)                  #get the url 
                soup=BeautifulSoup(response.text,'html.parser')     #output is parsed to get required readable content 
                table=soup.find("div",attrs={'id':'quote-summary'})  # scrape the table name
                pc=table.find("td",attrs={'data-test':"PREV_CLOSE-value"}).get_text()   # scrape Previous Close value
                op=table.find("td",attrs={'data-test':"OPEN-value"}).get_text()         # scrape Open value
                vl=table.find("td",attrs={'data-test':"TD_VOLUME-value"}).get_text()    # scrape Volume value
                mc=table.find("td",attrs={'data-test':"MARKET_CAP-value"}).get_text()   # scrape Market Cap value
                csv_writer.writerow([types.get(t),t,fn,pc,op,vl,mc])        # write all values in the csv file in required format
                
        count=count+1                        #increment the counter to point to next type
        
csv_file.close()                            # close the file after writing 

with open('stocks.csv','r',newline='') as csv_file:  # open the file and display companies belonging to the "Most Actives"
    csv_reader=csv.reader(csv_file)
    print("Most Actives")
    for i in csv_reader:
        if(i[0]=="Most Actives"):
            print(i[1]+" "+i[2])
        
csv_file.close()
print("\n")
with open('stocks.csv','r',newline='') as csv_file:  # open the file and display companies belonging to the "Gainers" 
    csv_reader=csv.reader(csv_file)
    print("Gainers")
    for i in csv_reader:
        if(i[0]=="Gainers"):
            print(i[1]+" "+i[2])
        
csv_file.close()
print("\n")

with open('stocks.csv','r',newline='') as csv_file:   # open the file and display companies belonging to the "Losers"
    csv_reader=csv.reader(csv_file)
    print("Losers")
    for i in csv_reader:
        if(i[0]=="Losers"):
            print(i[1]+" "+i[2])
        
csv_file.close()
print("\n")
with open('stocks.csv','r',newline='') as csv_file: # open the stock.csv file 
    csv_reader=csv.reader(csv_file)
    inp=input("Which stock are you interested in:")  # user is prompted to enter the comapnies name to display required data
    for i in csv_reader:  
        if(inp.upper()==i[1]):
            print(i[1]+" "+i[2])
            print("OPEN:",i[4])
            print("PREV CLOSE:",i[3])
            print("VOLUME:",i[5])
            print("MARKET CAP:",i[6])
            break
    else:
        print("No company found")
        
csv_file.close()                                  # the file is closed
```
